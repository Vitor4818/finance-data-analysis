```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
df_principal = pd.read_excel('/content/analises.xlsx', sheet_name='Principal')
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel("/content/analises.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/analises.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/analises.xlsx", sheet_name="Chatgpt")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct']/100
df_principal
```

```python
df_principal['valor_inicial'] = df_principal['valor_final']/(df_principal['var_pct']+1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final']-df_principal['valor_inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtde_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x> 0 else ('Desceu' if x<0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Ano de Fundação':'ano_fundacao'}).copy()
df_principal = df_principal.rename(columns={'Idade Aproximada (anos)':'idade'}).copy()
df_principal
```

```python
df_principal["idade"] = pd.to_numeric(df_principal["idade"], errors="coerce").fillna(0).astype(int)

df_principal["cat_idade"] = df_principal["idade"].apply(lambda x: "Entre 0 a 50 anos" if x <= 50 else "Entre 50 a 100 anos" if x <= 100 else "Mais de 100 anos")
df_principal
```

```python
#Pegando o maior valor de variação em reais
maior = df_principal['Variacao_rs'].max()

#Pegando o menor valor de variação em reais
menor = df_principal['Variacao_rs'].min()

#Calculando a média de variações em reais
media = df_principal['Variacao_rs'].mean()

#Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Função para formatar os valores do eixo X como reais
def formatar_reais(x, pos):
    return f'R${x:,.0f}'  # Formato R$ com separação de milhares

# Ordenando os dados do maior para o menor
df_analise_segmento = df_analise_segmento.sort_values(by='Variacao_rs', ascending=True).reset_index(drop=True)

# Criando o gráfico de barras horizontais
fig, ax = plt.subplots(figsize=(15, 6))
ax.barh(df_analise_segmento['Segmento'], df_analise_segmento['Variacao_rs'], color='blue')

# Adicionando o título
plt.title('Variação do preço por data', fontsize=16)

# Adicionando a legenda
plt.legend(['Fechamento'])

# Configurando os rótulos dos eixos
plt.ylabel('Segmento', fontsize=14)  # Nome do eixo Y (segmentos)
plt.xlabel('Variação', fontsize=14)  # Nome do eixo X (Variação)

# Aplicando a formatação de moeda no eixo X
ax.xaxis.set_major_formatter(FuncFormatter(formatar_reais))

# Ajustando espaçamento para evitar sobreposição
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
import plotly.express as px

# Criando o gráfico de barras
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Resultado')

# Atualizando o eixo Y para mostrar valores em reais (R$)
fig.update_yaxes(tickformat="$.2f")  # Formato de moeda, sem o prefixo "R$", pois o Plotly usa o simbolo da moeda

# Atualizando o texto nas barras para mostrar como moeda (R$)
fig.update_traces(texttemplate='R$%{y:,.2f}', textposition='outside')  # Formatação de moeda com 2 casas decimais

# Exibindo o gráfico
fig.show()
```
